```python
sns.countplot(x="Gender",data=df)
sns.set(rc={"figure.figsize":(2,3)})
```

```python
ax=sns.countplot(x="Gender",data=df)
sns.set(rc={"figure.figsize":(2,3)})

for bars in ax.containers:
  ax.bar_label(bars)
```

the female have done more shopping


```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount")
sns.barplot(x="Gender",y="Amount" , data = sales_gen)
sns.set(rc={"figure.figsize":(2,3)})
```

```python
ax=sns.countplot(data=df,x="Age Group")
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

from above graphs we can see that most of the buyers are of age group between 26-35yrs female


```python
df.columns
```

```python
sales_state=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(29)
sns.set(rc={"figure.figsize":(30,15)})
sns.barplot(data=sales_state,x="State",y="Orders")
```

from above graphs we can see that most of the orders are from UttarPradesh,Maharashtra,karnatka respectively.



```python
df.groupby(["Marital_Status"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
ax = sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(2,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Marital_Status","Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(6,7)})
sns.barplot(data=sales_state,x="Marital_Status",y="Amount",hue="Gender")
```

from above graphs we can see that most of the buyers are married womans(0) and they have high purchasing power


```python
df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
ax = sns.countplot(data=df,x="Occupation")
sns.set(rc={"figure.figsize":(35,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(25,10)})
sns.barplot(data=sales_state,x="Occupation",y="Amount")
```

from above graph we can see that most ofthe buyers are working in IT,Aviation and Healthcare sector


```python
ax = sns.countplot(data=df,x="Product_Category")
sns.set(rc={"figure.figsize":(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_state=df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(25,10)})
sns.barplot(data=sales_state,x="Product_Category",y="Amount")
```

from above graphs we can see that most of the sold products are from Food,clothes,electronics gadgets.


```python
df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
```

```python
sales_state=df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(25,10)})
sns.barplot(data=sales_state,x="Product_ID",y="Orders")
```

CONCLUSION:
Married Women age-group 26-35yrs from UP, Maharashtra, Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food,Clothing and Electronics category
